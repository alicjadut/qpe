```python
from sparse_qpe import kappa_finder, abs_phase_difference
import numpy as np
```

```python
num_phases = 10
for seed in range(100):
    rng = np.random.RandomState(seed)
    phases = rng.uniform(0, 2 * np.pi, num_phases)
    prev_multiplier = rng.uniform(2, 50)
    eps = rng.uniform(1e-3, 1e-2)
    kappa = kappa_finder(phases, eps, prev_multiplier)
    print(seed, kappa)
    for index1, phase1 in enumerate(phases):
        for phase2 in phases[index1+1:]:
            print(eps, 4 * eps * (1 + kappa),
                  abs_phase_difference(phase1, phase2),
                  abs_phase_difference(phase1 * kappa * prev_multiplier,
                                   phase2 * kappa * prev_multiplier))
    print()
```
